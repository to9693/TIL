```python
lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
        
df['lat'] = lat
df['lng'] = lng
df.head()
```

### 3-10 네이버 영화 평점 기준 영화의 평점 변화 확인하기

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from urllib.request import urlopen
```

```python
page = urlopen('https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190924')
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
soup.find_all('div', 'tit5')
```

```python
soup.find_all('div','tit5')[0].a.string # a 는 태그 이름
```

```python
soup.find_all('div', 'tit5')[0].a.get_text() # string 으로 뽑아야 타입이 뷰티풀숩으로 유지됨
```

```python
soup.find_all('td', 'point')[0].string
```

```python
date = pd.date_range('2017-5-1', periods=100, freq='D')
date
```

```python
import urllib
from tqdm import tqdm_notebook
```

```python
movie_date =[]
movie_name = []
movie_point = []
```

```python
for today in tqdm_notebook(date):
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}'
    
    response = urlopen(html.format(date = urllib.parse.quote(today.strftime('%Y%m%d'))))
    
    soup = BeautifulSoup(response, 'html.parser')
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all("div", 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})

movie.head()
```

```python
movie['point'] = movie['point'].astype(float)
movie.info() # point 가 문자 형태로 되있어서 숫자로 바꿔 줌
```

```python
import numpy as np
```

```python
movie_unique = pd.pivot_table(movie, index = ['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending = False)
movie_best.head()
```

```python
tmp = movie.query('name == ["노무현입니다"]')
tmp.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')
plt.grid()
plt.show()
```
